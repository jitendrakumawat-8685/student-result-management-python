Student Result Management System

```python
students = []
def add_student():
    name = input("Enter the Student Name => ")
    
    Hindi = input(f"Enter the Hindi subject Markes =>")
    English = input(f"Enter the English subject Markes =>")
    Maths = input(f"Enter the Maths subject Markes =>")
    Physics = input(f"Enter the Physics subject Markes =>")
    Chemistry = input(f"Enter the Chemistry subject Markes =>")
    student = {
        "name": name,
        "Hindi": Hindi,
        "English": English,
        "Maths": Maths,
        "Physics": Physics,
        "Chemistry": Chemistry
    }
    students.append(student)
    print(f"Student {name} added successfully!")

def calculate_grade(average):
    if average >=90:
        return "A"
    elif average >=80:
        return "B"
    elif average >=70:
        return "C"
    elif average >=60:
        return "D"
    else:
        return "F"
    
def show_result():
    name = input("Enter the Student Name => ")
    for student in students:
        if student["name"].lower() == name.lower():
            total = int(student["Hindi"]) + int(student["English"]) + int(student["Maths"]) + int(student["Physics"]) + int(student["Chemistry"])
            average = total / 5
            grade = calculate_grade(average)
            status = "Pass" if grade != "F" else "Fail"
            print(f"\n---{student['name']} Result---")
            print(f"Hindi   : {student['Hindi']}")
            print(f"English : {student['English']}")
            print(f"Maths   : {student['Maths']}")
            print(f"Physics : {student['Physics']}")
            print(f"Chemistry : {student['Chemistry']}")
            print(f"Total Marks: {total}")
            print(f"Average Marks: {average}")
            print(f"Grade: {grade}")
            print(f"Status: {status}")
            return
        
    print(f"Student {name} not found!")

def top_performance():
    print("\n---Top Performance (Average Marks 75+)---")
    found = False
    for student in students:
        total = int(student["Hindi"]) + int(student["English"]) + int(student["Maths"]) + int(student["Physics"]) + int(student["Chemistry"])
        average = total / 5
        if average >= 75:
            print(f"{student['name']} - Average Marks: {average}")
            found = True
    if not found:
        print("No students found with average marks 75 or above.")
    
def save_to_file():
    try:
        with open("students.txt", "r") as f:
            existing_students = f.read()
    except FileNotFoundError:
        existing_students = ""

    with open("students.txt", "a") as f:
        f.write("=== Student Results ===\n\n")
        for student in students:
            if student["name"].lower() not in existing_students.lower() :
                total = int(student["Hindi"]) + int(student["English"]) + int(student["Maths"]) + int(student["Physics"]) + int(student["Chemistry"])
                average = total / 5
                grade = calculate_grade(average)
                status = "Pass" if grade != "F" else "Fail"
                f.write(f"---{student['name']} Result---\n")
                f.write(f"Hindi   : {student['Hindi']}\n")
                f.write(f"English : {student['English']}\n")
                f.write(f"Maths   : {student['Maths']}\n")
                f.write(f"Physics : {student['Physics']}\n")
                f.write(f"Chemistry : {student['Chemistry']}\n")
                f.write(f"Total Marks: {total}\n")
                f.write(f"Average Marks: {average:.2f}\n")
                f.write(f"Grade: {grade}\n")
                f.write(f"Status: {status}\n\n")
    print("Results saved to students.txt successfully!")

def main():
    while True:
        print("\n--- Student Result Management System ---")
        print("1. Add Student")
        print("2. Show Result")
        print("3. Top Performance")
        print("4. Save to File")
        print("5. Exit")
        
        choice = input("Enter your choice (1-5) => ")
        
        if choice == '1':
            add_student()
        elif choice == '2':
            show_result()
        elif choice == '3':
            top_performance()
        elif choice == '4':
            save_to_file()
        elif choice == '5':
            print("Exiting the program. Goodbye!")
            break
        else:
            print("Invalid choice! Please enter a number between 1 and 5.")

if __name__ == "__main__":
    main()

```
